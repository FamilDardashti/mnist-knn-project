```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import mnist  
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print(X_train.shape, X_test.shape)   
print(y_train.shape, np.unique(y_train))
print(X_train.min(), X_train.max())  
```

```python
plt.figure(figsize=(10, 4))
for i in range(10):
    idx = np.random.randint(0, len(X_train))  
    plt.subplot(2, 5, i+1)                    
    plt.imshow(X_train[idx], cmap="gray")     
    plt.title(f"Label: {y_train[idx]}")
    plt.axis("off")
plt.show()
```

```python
unique, counts = np.unique(y_train, return_counts=True)

for label, count in zip(unique, counts):
    print(f"Digit {label}: {count} samples")

plt.bar(unique, counts)
plt.xlabel("Digit")
plt.ylabel("Number of samples")
plt.title("Samples per class")
plt.show()
```

```python
plt.figure(figsize=(10, 4))

for digit in range(10):
    images_of_digit = X_train[y_train == digit]
    
    mean_image = np.mean(images_of_digit, axis=0)
    
    plt.subplot(2, 5, digit+1)
    plt.imshow(mean_image, cmap="gray", vmin=0, vmax=255)  #   0..255 
    plt.title(f"Mean {digit}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
